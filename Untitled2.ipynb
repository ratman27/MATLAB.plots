```python
##converting categorical data into numerical type because machine cannot understand mumerical data
```

```python
crop_dict = {
    'rice': 1,
    'maize': 2,
    'chickpea': 3,
    'kidneybeans': 4,
    'pigeonpeas': 5,
    'mothbeans': 6,
    'mungbean': 7,
    'blackgram': 8,
    'lentil': 9,
    'pomegranate': 10,
    'banana': 11,
    'mango': 12,
    'grapes': 13,
    'watermelon': 14,
    'muskmelon': 15,
    'apple': 16,
    'orange': 17,
    'papaya': 18,
    'coconut': 19,
    'cotton': 20,
    'jute': 21,
    'coffee': 22
}
crop['crop_num'] = crop['label'].map(crop_dict)
```

```python
crop['crop_num'].value_counts()
```

```python
# Check if 'label' column exists before dropping
if 'label' in crop.columns:
    crop.drop('label', axis=1, inplace=True)
else:
    print("Column 'label' has already been dropped.")
crop.head()
```
